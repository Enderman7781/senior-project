```python
import csv
import json
import os
```

```python
def write_to_csv(json_data, write_file):
    with open(write_file, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['VDID', 'LinkID', 'LaneID', 'LaneType', 'Speed', 'Occupancy', 'VehicleType', 'Volume', 'VehicleSpeed', 'Status', 'DataCollectTime'])

        for vdlive in json_data['VDLiveList']['VDLive']:
            VDID = vdlive['VDID']
            Status = vdlive['Status']
            DataCollectTime = vdlive['DataCollectTime']
            
            link_flows = vdlive['LinkFlows']['LinkFlow']
            # Handle the case where there is only one LinkFlow (not a list)
            if isinstance(link_flows, dict):
                link_flows = [link_flows]

            for link_flow in link_flows:
                LinkID = link_flow['LinkID']
                lanes = link_flow['Lanes']['Lane']
                # Handle the case where there is only one Lane (not a list)
                if isinstance(lanes, dict):
                    lanes = [lanes]

                for lane in lanes:
                    LaneID = lane['LaneID']
                    LaneType = lane['LaneType']
                    Speed = lane['Speed']
                    Occupancy = lane['Occupancy']
                    vehicles = lane['Vehicles']['Vehicle']
                    # Handle the case where there is only one Vehicle (not a list)
                    if isinstance(vehicles, dict):
                        vehicles = [vehicles]

                    for vehicle in vehicles:
                        VehicleType = vehicle['VehicleType']
                        Volume = vehicle['Volume']
                        VehicleSpeed = vehicle['Speed']
                        writer.writerow([VDID, LinkID, LaneID, LaneType, Speed, Occupancy, VehicleType, Volume, VehicleSpeed, Status, DataCollectTime])
```

```python
month = ['{0:02d}'.format(x) for x in range(1,11)]
day = ['{0:02d}'.format(x) for x in range(1,32)]

month_upper = [31,28,31,30,31,30,31,31,30,31,30,31]
print(month,day)

base_path = r'D:\Desktop\Code\專題\DATA\2023'
write_path = r'D:\Desktop\Code\專題\CSV_VD_DATA\2023'

for i in month:
    for j in day:
        if int(j) > month_upper[int(i)-1]:
            break
        for k in range(0,24):
            k = '{0:02d}'.format(k)
            for l in range(0,60,10):
                l = '{0:02d}'.format(l)
                json_file = base_path+i+j+'\\2023'+i+j+'_'+k+l+'.json'
                csv_file = write_path+i+j+'\\2023'+i+j+'_'+k+l+'.csv'
                
                os.makedirs(os.path.dirname(csv_file), exist_ok=True)
                if os.path.exists(json_file):
                    # Read the JSON data and write to CSV
                    with open(json_file, 'r', encoding='utf-8') as f:
                        json_data = json.load(f)
                    write_to_csv(json_data, csv_file)
                else:
                    print(f"File {json_file} does not exist. Skipping.")
        
```
